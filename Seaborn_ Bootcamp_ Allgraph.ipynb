```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,5))
```

#  1. Scatterplot

```python
sns.scatterplot(data=tips,x="total_bill",y="tip",hue="time",size="size",palette="deep")
plt.title("ScatterPlot of Total Bill vs Tip")
plt.show()
```

# 2. Line plot

```python
sns.lineplot(data=tips,x="size",y="total_bill",hue="sex",markers='o',ci=None)
plt.title("Line Plot of Total Bill vs size")
plt.show()
```

```python

sns.lineplot(data=tips,x="size",y="total_bill",hue="sex",markers='o',)
plt.title("Line Plot of Total Bill vs size")
plt.show()
```

# 3. Bar plot

```python
sns.barplot(data=tips,x="day",y="total_bill",hue="sex",palette='muted')
plt.title("Bar Plot of Total Bill vs Day")
plt.show()
```

# 4. Box plot

```python
tips.columns
```

```python
sns.boxplot(data=tips,x="day",y="tip",hue="smoker",palette='Set1')
plt.title("BoxPlot of Tips by Day and smoker status")
plt.show()
```

# 5. Violin plot

```python
sns.violinplot(data=tips,x="day",y="total_bill",hue="time",split=True,)
plt.title("Violin Plot of Total bill by Day and time")
plt.show()
```

```python
tips.columns
```

# 6. count plot

```python
sns.countplot(data=tips,x="day",hue="smoker",palette='bright')
plt.title("Count Plot of Day by smoker status")
plt.show()
```

# 7. regression plot

```python
sns.regplot(data=tips,x="total_bill",y="tip",scatter_kws={'s':50},line_kws={'color':'red'})
plt.title("Regression Plot of Total Bill vs Tip")
plt.show()
```

# 8. Histogram

```python
sns.histplot(data=tips,x="total_bill",bins=10,kde=True,color='red')
plt.title("Histogram of Total Bill with kde")
plt.show()
```

# 9.pairplot

```python
sns.pairplot(tips,hue='sex',vars=["total_bill","tip","size"],palette='bright')
plt.title("pair plot: numerberic variables by gender", y=1.02)
plt.show()
```

# 10. cat plot

```python
sns.catplot(data=tips,x="day",y="tip",hue="sex",kind="point",palette="bright")
plt.title("catplot(point): tip by day and gender")
plt.show()
```

# 11. Jiont plot

```python
sns.jointplot(data=tips,x="total_bill",y="tip",kind="scatter",hue="smoker",color="purple",palette="coolwarm")
plt.title("JointPlot: Total Bill vs Tip",y=1.02)
plt.show()
```

# 12.facetgrid


```python
g = sns.FacetGrid(tips,col="time",row="smoker",margin_titles=True).map(sns.histplot,"total_bill")
g
```

# 13. strip plot

```python
sns.stripplot(data=tips,x="day",y="tip",hue="sex",jitter=True,palette="Set1")
plt.title("strip plot: tips by day and gender")
plt.show()
```

# 14.kde plot

```python
sns.kdeplot(data=tips,x="total_bill",hue="sex",fill=True,palette="tab10")
plt.title("KDE Plot: Total Bill density by Gender")
plt.show()
```
